## 展示出前若干个样本

```python
import numpy as np
import pandas as pd
from time import time
from IPython.display import display
%matplotlib inline

data = pd.read_csv("voice.csv")
display(data.head(n=3))

n_records = len(data)
n_male = len(data[data["label"]=="male"])
n_female = len(data[data["label"]=="female"])
male_percent = n_male*100.0/n_records
female_percent = n_female*100.0/n_records

print "数据集中的总样本数: {}".format(n_records)
print "男性语音样本数: {}".format(n_male)
print "男性语音样本数: {}".format(n_female)
print "男性样本和女性样本的占比分别为: {}%， {}%".format(male_percent, female_percent)
```

```python
# 将数据切分成特征和对应的标签
label_raw = data['label']
features_raw = data.drop('label', axis = 1)

# 将'label'编码成数字值
label_mapping = {'male':1, 'female':0}
label = label_raw.map(label_mapping)
```

## 画出不同性别样本个数的柱状图

```python
# 画出不同性别样本个数的柱状图

import matplotlib.pyplot as plt
plt.figure(figsize = (6,5))
pos=[0,1]#横轴id
data['label'].value_counts().plot.bar(color = '#00A0A0')
plt.title("The number of each label", fontsize=14)
# plt.xlabel("Sex", fontsize=15)
plt.ylabel("Numbers", fontsize=14)
plt.show()
```

```python
# 几个特征在不同性别的分布情况

fig_meanfreq = plt.figure(figsize = (12,24))
male_mf = data[data['label']=='male']['meanfreq']
female_mf = data[data['label']=='female']['meanfreq']

male_mfun = data[data['label']=='male']['meanfun']
female_mfun = data[data['label']=='female']['meanfun']

male_md = data[data['label']=='male']['meandom']
female_md = data[data['label']=='female']['meandom']

male_mode = data[data['label']=='male']['dfrange']
female_mode = data[data['label']=='female']['dfrange']

ax = fig_meanfreq.add_subplot(421)
ax.boxplot([male_mf*1000, female_mf*1000], showfliers=False)
ax.set_xticklabels(['Male', 'Female'])
ax.set_ylim((100, 260))
ax.set_title("Meanfreq", fontsize = 14)
ax.set_ylabel("Frequency(Hz)", fontsize = 14)
ax.grid(color='grey', linestyle='--', axis='y')

ax = fig_meanfreq.add_subplot(423)
ax.boxplot([male_mfun*1000, female_mfun*1000], showfliers=False)
ax.set_xticklabels(['Male', 'Female'])
ax.set_ylim((60, 220))
ax.set_title("Meanfun", fontsize = 14)
ax.set_ylabel("Frequency(Hz)", fontsize = 14)
ax.grid(color='grey', linestyle='--', axis='y')

ax = fig_meanfreq.add_subplot(425)
ax.boxplot([male_md*1000, female_md*1000], showfliers=False)
ax.set_xticklabels(['Male', 'Female'])
ax.set_ylim((0, 2700))
ax.set_title("Meandom", fontsize = 14)
ax.set_ylabel("Frequency(Hz)", fontsize = 14)
ax.grid(color='grey', linestyle='--', axis='y')

ax = fig_meanfreq.add_subplot(427)
ax.boxplot([male_mode*1000, female_mode*1000], showfliers=False)
ax.set_xticklabels(['Male', 'Female'])
# ax.set_ylim((0, 300))
ax.set_title("Dfrange", fontsize = 14)
ax.set_ylabel("Frequency(Hz)", fontsize = 14)
ax.grid(color='grey', linestyle='--', axis='y')

# hist = plt.figure(figsize = (12,24))

hist = fig_meanfreq

h = hist.add_subplot(422)
h.set_title("Histogram of Meanfreq", fontsize = 14)
h.set_xlabel("Frequency(Hz)", fontsize = 14)
h.set_ylabel("Number of Records", fontsize = 14)
h.hist(data['meanfreq']*1000, bins = 25, color = '#00A0A0')
h.grid(color='grey', linestyle='--', axis='y')
h.set_ylim((0, 500))
h.set_xlim((50, 250))

h = hist.add_subplot(424)
h.set_title("Histogram of Meanfun", fontsize = 14)
h.set_xlabel("Frequency(Hz)", fontsize = 14)
h.set_ylabel("Number of Records", fontsize = 14)
h.hist(data['meanfun']*1000, bins = 25, color = '#00A0A0')
h.grid(color='grey', linestyle='--', axis='y')
h.set_ylim((0, 300))
h.set_xlim((50, 250))

h = hist.add_subplot(426)
h.set_title("Histogram of Meandom", fontsize = 14)
h.set_xlabel("Frequency(Hz)", fontsize = 14)
h.set_ylabel("Number of Records", fontsize = 14)
h.hist(data['meandom']*1000, bins = 25, color = '#00A0A0')
h.grid(color='grey', linestyle='--', axis='y')
h.set_ylim((0, 500))
h.set_xlim((0, 3000))

h = hist.add_subplot(428)
h.set_title("Histogram of Dfrange", fontsize = 14)
h.set_xlabel("Frequency(Hz)", fontsize = 14)
h.set_ylabel("Number of Records", fontsize = 14)
h.hist(data['dfrange']*1000, bins = 25, color = '#00A0A0')
h.grid(color='grey', linestyle='--', axis='y')
h.set_ylim((0, 700))
h.set_xlim((0, 22000))

plt.show()
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import MinMaxScaler

lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(features_raw, label)
feature_new = lda.transform(features_raw)

scaler = MinMaxScaler()
feature_new = scaler.fit_transform(feature_new)
# print(feature_new[:10])
plt.figure(figsize = (8,7))
plt.scatter(feature_new, [1]*len(feature_new), marker='o', c=label)
plt.yticks([])
plt.xlim((0, 1))
plt.grid(color='grey', linestyle='--', axis='x')

plt.xlabel('Value of New Feature', fontsize=14)
plt.title('LDA', fontsize=14)
plt.show
```

## Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# 将'features'和'label'数据切分成训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(features_raw, label, 
                                test_size=0.2, random_state=0, stratify=label)
rf = RandomForestClassifier(oob_score=False)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from matplotlib import cm

fs = zip(data.columns,rf.feature_importances_)
fs_sorted = sorted(fs, key=lambda z: z[1], reverse=False)

label, x = [_[0] for _ in fs_sorted], [_[1] for _ in fs_sorted]

plt.figure(figsize = (10,8))
idx = np.arange(len(x))
color = cm.jet(np.array(x)/max(x))
color = np.random.random(size=(20, 4))
plt.barh(idx, x, color=color)
plt.yticks(idx+0.4,label)
plt.grid(color='grey', linestyle='--', axis='x')
plt.xlim((0, 0.4))
 
plt.xlabel('Score', fontsize=14)
plt.title('Feature Importances', fontsize=14)
 
plt.show()
```
